```python
from datasets import load_dataset
import jsonlines

def process_dataset(dataset_name, dataset_link):
    dataset = load_dataset(dataset_name)

    with jsonlines.open('medical_transcriptions_full.jsonl', mode='w') as writer:
        for i, item in enumerate(dataset['train']):
            if isinstance(item['keywords'], (list, tuple)):
                keywords = ', '.join(item['keywords'])
            elif isinstance(item['keywords'], str):
                keywords = item['keywords']
            else:
                keywords = str(item['keywords'])

            text = f"Description: {item['description']}\n" \
                   f"Medical Specialty: {item['medical_specialty']}\n" \
                   f"Sample Name: {item['sample_name']}\n" \
                   f"Transcription: {item['transcription']}\n" \
                   f"Keywords: {keywords}"

            entry = {
                'meta': {
                    'dataset_link': dataset_link,
                    'dataset_name': dataset_name,
                    'id': i
                },
                'text': text
            }
            writer.write(entry)

dataset_name = "tchebonenko/MedicalTranscriptions"
dataset_link = "https://huggingface.co/datasets/tchebonenko/MedicalTranscriptions"
process_dataset(dataset_name, dataset_link)

print("JSONL file 'medical_transcriptions_full.jsonl' has been created.")
```
