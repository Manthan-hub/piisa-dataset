```python
from datasets import load_dataset
import jsonlines
```

```python
dataset = load_dataset("rungalileo/medical_transcription_40")

with jsonlines.open('medical_transcription_40.jsonl', mode='w') as writer:
    for i, item in enumerate(dataset['train']):
        writer.write({
            'meta': {
                'dataset_link': 'https://huggingface.co/datasets/rungalileo/medical_transcription_40',
                'dataset_name': 'Medical Transcription 40',
                'id': i
            },
            'text': item['text']
        })
```
