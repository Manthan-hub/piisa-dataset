```python
from datasets import load_dataset
import jsonlines
```

```python
def process_dataset(dataset_name, dataset_link):
    dataset = load_dataset(dataset_name)
    for i, item in enumerate(dataset['train']):
        yield {
            'meta': {
                'dataset_link': dataset_link,
                'dataset_name': dataset_name,
                'id': i
            },
            'text': item['text']
        }

with jsonlines.open('combined_medical_transcription.jsonl', mode='w') as writer:
    for item in process_dataset("rungalileo/medical_transcription_4",
                                "https://huggingface.co/datasets/rungalileo/medical_transcription_4"):
        writer.write(item)

    for item in process_dataset("rungalileo/medical_transcription_40",
                                "https://huggingface.co/datasets/rungalileo/medical_transcription_40"):
        writer.write(item)
```
