```python
from datasets import load_dataset
import jsonlines
```

```python
def process_dataset(dataset_name, dataset_link):
    dataset = load_dataset(dataset_name)
    for i, item in enumerate(dataset['train']):
        # Concatenate relevant fields
        text = f"{item['ehr_text']} ICD Code: {item['icd_code']}"
        yield {
            'meta': {
                'dataset_link': dataset_link,
                'dataset_name': dataset_name,
                'id': i
            },
            'text': text
        }

# Process the dataset
with jsonlines.open('synth_ehr_icd10.jsonl', mode='w') as writer:
    for item in process_dataset("generative-technologies/synth-ehr-icd10-llama3-format",
                                "https://huggingface.co/datasets/generative-technologies/synth-ehr-icd10-llama3-format"):
        writer.write(item)
```
