```python
from datasets import load_dataset
import jsonlines

def process_dataset(dataset_name, dataset_link):
    dataset = load_dataset(dataset_name)

    with jsonlines.open('ehrcomplete_icdfiltered.jsonl', mode='w') as writer:
        for i, item in enumerate(dataset['train']):
            text = f"Medical Text: {item['TEXT']}\nICD9 Codes: {', '.join(item['ICD9_CODE'])}"

            entry = {
                'meta': {
                    'dataset_link': dataset_link,
                    'dataset_name': dataset_name,
                    'id': i
                },
                'text': text
            }
            writer.write(entry)

dataset_name = "ricardosantoss/ehrcomplete_icdfiltered"
dataset_link = "https://huggingface.co/datasets/ricardosantoss/ehrcomplete_icdfiltered"
process_dataset(dataset_name, dataset_link)

print("JSONL file 'ehrcomplete_icdfiltered.jsonl' has been created.")
```
